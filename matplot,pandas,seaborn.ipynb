```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate some random data
import numpy as np
data = np.random.randn(1000)  # 1000 data points sampled from a normal distribution

# Create a histogram with Seaborn (automatic binning)
sns.histplot(data,kde=False, bins=20)  # You can change the number of bins here
plt.title('Histogram with 20 Bins')
plt.show()
```

```python
bin_edges = [-3, -2, -1, 0, 1, 2, 3]
sns.histplot(data, bins=bin_edges)
```
